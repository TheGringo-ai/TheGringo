```python
import os
import hashlib
import shutil
from collections import defaultdict
from google.colab import drive

# Mount Google Drive
drive.mount('/content/drive')

# Define root folder
root_folder = "/content/drive/MyDrive/BookProject"
duplicate_folder = os.path.join(root_folder, "_duplicates")

# Create duplicate folder if it doesn't exist
os.makedirs(duplicate_folder, exist_ok=True)

# Function to hash file content
def get_file_hash(filepath):
    sha256 = hashlib.sha256()
    try:
        with open(filepath, 'rb') as f:
            while chunk := f.read(8192):
                sha256.update(chunk)
        return sha256.hexdigest()
    except Exception as e:
        print(f"âŒ Error hashing {filepath}: {e}")
        return None

# Step 1: Walk through all files and group by hash
hash_map = defaultdict(list)

for dirpath, _, filenames in os.walk(root_folder):
    for fname in filenames:
        full_path = os.path.join(dirpath, fname)
        if duplicate_folder in full_path:
            continue  # Skip already moved files

        file_hash = get_file_hash(full_path)
        if file_hash:
            hash_map[file_hash].append(full_path)

# Step 2: Identify duplicates and move them
moved_count = 0

for hash_val, paths in hash_map.items():
    if len(paths) > 1:
        # Keep the first one, move others
        original = paths[0]
        duplicates = paths[1:]
        for dup_path in duplicates:
            try:
                filename = os.path.basename(dup_path)
                target_path = os.path.join(duplicate_folder, filename)

                # If file already exists in _duplicates, rename
                if os.path.exists(target_path):
                    base, ext = os.path.splitext(filename)
                    target_path = os.path.join(duplicate_folder, f"{base}_DUP{ext}")

                shutil.move(dup_path, target_path)
                moved_count += 1
                print(f"â¡ï¸ Moved duplicate: {dup_path} â†’ {target_path}")
            except Exception as e:
                print(f"âŒ Failed to move {dup_path}: {e}")

print(f"\nâœ… Done. Moved {moved_count} duplicate files to: {duplicate_folder}")
```

```python
import os
import shutil
from google.colab import drive

# Mount Google Drive
drive.mount('/content/drive')

# Define base path
base_path = "/content/drive/MyDrive/BookProject"

# Define keyword groups and destination folders
routing_rules = {
    "drafts": ["gringo", "incident", "shift"],
    "characters": ["mercedes", "chencho", "jone"],
    "research": ["plant", "procedure", "law", "safety"],
    "promo_material": ["blurb", "title", "cover", "promo"]
}

# Create folders if missing
for folder in routing_rules.keys():
    os.makedirs(os.path.join(base_path, folder), exist_ok=True)

# Walk all files (excluding _duplicates/)
moved_count = 0
for dirpath, _, filenames in os.walk(base_path):
    if "_duplicates" in dirpath:
        continue  # Skip duplicates folder

    for fname in filenames:
        full_path = os.path.join(dirpath, fname)

        # Skip if already in destination folder
        if any(folder in full_path for folder in routing_rules.keys()):
            continue

        # Check keyword match
        fname_lower = fname.lower()
        matched = False
        for folder, keywords in routing_rules.items():
            if any(kw in fname_lower for kw in keywords):
                dest_path = os.path.join(base_path, folder, fname)

                # Auto-rename if conflict
                if os.path.exists(dest_path):
                    base, ext = os.path.splitext(fname)
                    dest_path = os.path.join(base_path, folder, f"{base}_SORTED{ext}")

                shutil.move(full_path, dest_path)
                print(f"ğŸ“‚ Moved: {fname} â†’ /{folder}/")
                moved_count += 1
                matched = True
                break

        if not matched:
            print(f"ğŸŸ¡ Skipped (no keyword match): {fname}")

print(f"\nâœ… Sorting complete. {moved_count} files moved into classified folders.")
```
